## Graphing LP

This section would illustrate the Graphing Method to solve the Linear Programming. We will use Python to help us graphing the Linear Programming.

Let us consider following Linear Programming!

**Decision Variables**

Our objective is to determine the optimal values of $x$ and $y$ that maximize the objective function of this LP.

**Objective function**

We want to find the maximum solution to the objective function: \
$Z = 4x+3y$.

**Constraints**

The objective function is subject to the following constraints: \
$x \geq 0 $ \
$y \geq 2 $ \
$2y \leq 25-x$ \
$4y \geq 2x-8$ \
$y \leq 2x-5$

**Step 1**: Importing Libraries

We call two libraries `matplotlib` and `numpy` to plot these equations. You could ignore this if you already imported the libraries earlier

```python
# if you already import it earlier, then, you do not need to run this code
import matplotlib.pyplot as plt
import numpy as np
```

**Step 2**: Define the Dataset of each Constraint Lines

We need to define a XY dataset of each equation by defining X and calculate their corresponding Y based on the equation.

We use `np.linspace(start point, end point, number of point)` to define a range of array. In this exercise, it asked to plot the equations from $[0,20]$ with 2000 data points

```python
# Construct dataset for each constraint lines
# x > 0
x = np.linspace(0, 20, 2000)
# y >= 2
y1 = (x*0) + 2
# 2y <= 25 - x
y2 = (25-x)/2.0
# 4y >= 2x - 8
y3 = (2*x-8)/4.0
# y <= 2x - 5
y4 = 2 * x -5
```

**Step 3**: Plot

Let us plot our input dataset simply by using `plt.plot()` function.

```python
plt.plot(x, y1, label='$y\geq2$')
plt.plot(x, y2, label='$2y\leq25-x$')
plt.plot(x, y3, label='$4y\geq 2x - 8$')
plt.plot(x, y4, label='$y\leq 2x-5$')
plt.xlim((0, 20))
plt.ylim((0, 20))
plt.xlabel("x")
plt.ylabel("y")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

**Draw the Feasible Region!**

Then, we fill the feasible region in accordance to the constrains inequalities:

$ y \geq 2 $, feasible region above the line

$2y\leq25-x$, feasible region below the line

$4y\geq 2x - 8$, feasible region above the line

$y\leq 2x-5$, feasible region below the line

Then, the maximum y value of feasible region (upper bound) is `minimum (y2, y4)` and the minimum y value of feasible region (lower bound) is `maximum(y1,y3)`

```python
# Plot the equation lines
plt.plot(x, y1, label='$y\geq2$')
plt.plot(x, y2, label='$2y\leq25-x$')
plt.plot(x, y3, label='$4y\geq 2x - 8$')
plt.plot(x, y4, label='$y\leq 2x-5$')
plt.xlim((0, 20))
plt.ylim((0, 20))
plt.xlabel("x")
plt.ylabel("y")

# Plot the feasible region
y5 = np.minimum(y2, y4) #maximum value of feasible region (upper bound)
y6 = np.maximum(y1, y3) #minimum value of feasible region (lower bound)
plt.fill_between(x, y5, y6, where=y5>y6, color='grey', alpha=0.5)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

Our solution lies somewhere in the grey feasible region in the graph above.\
\
It has been proven that the minima and maxima of linear programming problems lie at the vertices of the feasible region. In this example, there are only 4 corners to our feasible region, so we can find the solutions for each corner to find our maximum. \
\
We need to examine the value of the objective function $Z$ at the 4 corners, which are the intersections of the following equations:

| Corner | Line 1      | Line 2      |
|--------|-------------|-------------|
|1       | y  2       | 4y  2x  8 |
|2       | 2y  25  x | y  2x  5  |
|3       | 2y  25  x | 4y  2x  8 |
|4       | y  2       | y  2x  5  |

Once again, our objective is to maximize:
$Z=4x+3y$, thus we calculate $Z$ for each corner:

**Corner 1**:\
$y2$ and $4y2x8$\
$2=\frac{2x8}{4}$ and $x=8, y=2$ and $Z=(48+32)=38$

**Corner 2**:\
$2y25x$ and $y2x5$ \
$\frac{25x}{2}=2x5$ \
$x=7, y=9$ and $Z=(47+39)=55$

**Corner 3**:\
$2y25x$ and $4y\geq2x8$ \
$\frac{25x}{2}=\frac{2x8}{4}$ \
$x=14.5, y=5.25$ and $Z=(414.5+35.25)=73.75$

**Corner 4**:\
$y2$ and $y2x5$\
$2=2x5$ and $x=3.5, y=2$ then $Z=(43.5+32)=20$


Thus the maximum $Z$ is for $x=14.5, y=5.25$ with $Z=73.75$

## PuLP Direct Solver

Instead of using Graphing Method, we could solve LP using available direct solver.

In this tutorial, we will use PuLP packages/libraries to solve the LP directly. PuLP is one of python packages for solving LP problems. It assists us with constructing and solving linear programming problems. You could read their [manual here](https://coin-or.github.io/pulp/).

We will solve the previous LP using PuLP

$max \ Z = 4x+3y$

$s.t.$

$x \geq 0 $ \
$y \geq 2 $ \
$2y \leq 25-x$ \
$4y \geq 2x-8$ \
$y \leq 2x-5$


**Step 1**: Importing Libraries

We need to call library `pulp`.

**Note:** Your local computer might not have `pulp` installed yet. You need to install `pulp` first before import the library!




```python
## MAKE SURE TO INSTALL PULP BEFORE IMPORTING
## Ignore this if you already install it

!pip install pulp
```

```python
## After you install pulp, you could import it
import pulp
```

**Step 2**: Define the LP problem

Define an LP problem class using "My LP problem" convention. If we seeking to maximize an objective function we use LpMaximize, otherwise LpMinimize.

In our exercise, the objective function is to maximize. Hence, we define the LP problem as LpMaximize.

```python
my_lp_problem = pulp.LpProblem("My LP Problem", pulp.LpMaximize)
```

**Step 3**: Define Decision Variables

We use LpVariable class to model our decision variables. Upper bounds are denoted with the upBound parameter and lower bound with the lowBound papameter.

In our exercise, we have two decision variables, x and y, where:

$x \geq 0 $ \
$y \geq 2 $ \

We can treat these inequalities as lowerbound (the lowest possible value of x and y).

```python
x = pulp.LpVariable('x', lowBound=0, cat='Continuous')
y = pulp.LpVariable('y', lowBound=2, cat='Continuous')
```

**Step 4**: Define Objective Function

In PuLP, we need to define objective function before the constraints.

The objective function is added using the **+=** operator to the model.

In our exercise, the objective function is:

$Z = 4x+3y$

```python
# Objective function
my_lp_problem += 4 * x + 3 * y, "Z"
```

**Step 4**: Define Constraints

The constraints are also added using the **+=** operator to the model.

The remaining constraints:

$2y \leq 25-x$ \
$4y \geq 2x-8$ \
$y \leq 2x-5$



```python
# Constraints
my_lp_problem += 2 * y <= 25 - x
my_lp_problem += 4 * y >= 2 * x - 8
my_lp_problem += y <= 2 * x - 5
```

Let's take a look at the way that the problem was coded:

```python
my_lp_problem
```

**Step 5**: Solve it

PuLP is supporting solvers that are open source (default is [Coin Brunch and Cut](https://projects.coin-or.org/Cbc)) as well as proprietary, commercial solvers (like [Gurobi](https://www.gurobi.com) and [CPLEX](https://www.ibm.com/products/ilog-cplex-optimization-studio)).

We call the solver to find the solution to our problem:

```python
my_lp_problem.solve()
pulp.LpStatus[my_lp_problem.status]
```

In PuLP, there are several potential status.

The LpStatus outcome could be:
* optimal
* not solved
* infeasible (no feasible solution)
![image.png](https://raw.githubusercontent.com/ekontou/CEE201/a326cf4486012a34b399d7d8be480effa6836eaa/LPGraphing/Infeasible.png)
* unbounded (constraints are not bounded and solution tend to approach infinity when maximizing)
![image.png](https://raw.githubusercontent.com/ekontou/CEE201/a326cf4486012a34b399d7d8be480effa6836eaa/LPGraphing/Unbounded.png)
* undefined (optimal solution may exist but was not found)

```python
#Print decision variables
print(" x = {}".format(x.varValue))
print(" y = {}".format(y.varValue))
```

```python
#Print objective function value (max Z)
print(pulp.value(my_lp_problem.objective))
```

**We arrived to the same conclusion - compare the solution to the graphical one- using a python solver.**

**Exercise 1**

$\min_{x_1,x_2} 7 x_1 - 5 x_2$

$-2 x_1 - x_2 \leq 4 $ \
$ 10 x_1 + 3 x_2 \leq 30$ \
$ x_1 + 2 x_2 \geq 10 $ \

```python
## Define Problem
my_lp_problem = pulp.LpProblem("My LP Problem", pulp.LpMinimize)

# Define Decision Variables
x1 = pulp.LpVariable('x1', lowBound=0, cat='Continuous')
x2 = pulp.LpVariable('x2', lowBound=0, cat='Continuous')

# Define Objective Function
my_lp_problem += 7 * x1 - 5 * x2, "Z"

# Define Remaining Constraints
my_lp_problem += -2*x1 -x2 <= 4
my_lp_problem += 10*x1 + 3 *x2 <= 30
my_lp_problem += x1 + 2 * x2 >= 10
```

```python
my_lp_problem
```

```python
my_lp_problem.solve()
pulp.LpStatus[my_lp_problem.status]
```

```python
#Print decision variables
print(" x1 = {}".format(x1.varValue))
print(" x2 = {}".format(x2.varValue))

#Print objective function value (max Z)
print(pulp.value(my_lp_problem.objective))
```

**Exercise 2**

$\min_{x_1,x_2, x_3} 10 x_1 - 3 x_2 + 5 x_3$

$3 x_1 + 4 x_2 + x_3 \leq 100 $ \
$ 6 x_1 - 4 x_2 + 5 x3 \geq 31$ \

$x_1, x_3 \geq 0 $ \

```python
## Define Problem
my_lp_problem = pulp.LpProblem("My LP Problem", pulp.LpMinimize)

# Define Decision Variables
x1 = pulp.LpVariable('x', lowBound=0, cat='Continuous')
x2 = pulp.LpVariable('y', lowBound=2, cat='Continuous')
x3 = pulp.LpVariable('u', lowBound=0, cat='Continuous')

# Define Objective Function
my_lp_problem += 10 * x1 - 3 * x2 + 5*x3, "Z"

# Define Remaining Constraints
my_lp_problem += 3*x1 + 4*x2 + x3 <= 100
my_lp_problem += 6*x1 - 4 *x2 + 5*x3 >= 31
my_lp_problem += 5*x1 + 2*x2 +2*x3 >= 13
```

```python
my_lp_problem
```

```python
my_lp_problem.solve()
pulp.LpStatus[my_lp_problem.status]
```

```python
#Print decision variables
print(" x1 = {}".format(x1.varValue))
print(" x2 = {}".format(x2.varValue))
print(" x3 = {}".format(x3.varValue))

#Print objective function value (max Z)
print(pulp.value(my_lp_problem.objective))
```
